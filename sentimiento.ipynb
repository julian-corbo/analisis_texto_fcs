```python
from pysentimiento import create_analyzer
```

```python
analyzer = create_analyzer(task = "sentiment",  lang = "es")
```

```python
import pyreadr
```

```python
casarosada = pyreadr.read_r('/content/casarosada_final.Rdata')
```

```python
casarosada = casarosada['casarosada']
```

```python
from pysentimiento.preprocessing import preprocess_tweet
```

```python
def preprocess_text(texto):
    """Preprocesses a single tweet."""
    return preprocess_tweet(texto)
```

```python
casarosada['text_proc'] = casarosada['discurso'].astype(str).apply(preprocess_text)
```

```python
import pandas as pd
```

```python
print(casarosada['text_proc'].head())
```

```python
def analizar_sentimiento(texto):
    sentiment = analyzer.predict(texto).output
    return sentiment
```

```python
casarosada['text_proc'] = casarosada['text_proc'].astype(str)
casarosada['sentimiento'] = casarosada['text_proc'].apply(analizar_sentimiento)
```

```python
casarosada.to_csv("/content/casarosada_final_sentiment.csv", index=False)
```

```python
analyzerh = create_analyzer(task = "hate_speech", lang = "es")
```

```python
def analizar_hate(texto):
    hate = analyzerh.predict(texto).output
    return hate
```

```python
casarosada['text_proc'] = casarosada['text_proc'].astype(str)
casarosada['hate'] = casarosada['text_proc'].apply(analizar_hate)
```

```python
casarosada.to_csv("/content/casarosada_hate.csv", index=False)
```
